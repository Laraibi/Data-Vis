```python
# Filter all matches involving Morocco
morocco_matches = results[
    (results["home_team"] == "Morocco") | (results["away_team"] == "Morocco")
].copy()

morocco_matches.head()
```

```python
morocco_matches.count()
```

```python
# Identify if Morocco played home or away
morocco_matches["morocco_is_home"] = morocco_matches["home_team"] == "Morocco"

# Opponent
morocco_matches["opponent"] = morocco_matches.apply(
    lambda row: row["away_team"] if row["home_team"] == "Morocco" else row["home_team"],
    axis=1
)

# Morocco goals / opponent goals
morocco_matches["morocco_goals"] = morocco_matches.apply(
    lambda row: row["home_score"] if row["home_team"] == "Morocco" else row["away_score"],
    axis=1
)

morocco_matches["opponent_goals"] = morocco_matches.apply(
    lambda row: row["away_score"] if row["home_team"] == "Morocco" else row["home_score"],
    axis=1
)

# Match result for Morocco
morocco_matches["morocco_result"] = morocco_matches.apply(
    lambda row: "Win" if row["morocco_goals"] > row["opponent_goals"]
    else ("Loss" if row["morocco_goals"] < row["opponent_goals"] else "Draw"),
    axis=1
)

# Goal difference
morocco_matches["goal_diff"] = morocco_matches["morocco_goals"] - morocco_matches["opponent_goals"]

# Decade
morocco_matches["decade"] = (morocco_matches["year"] // 10) * 10

morocco_matches.head()
```

```python
print("Total matches:", len(morocco_matches))
print("Wins:", (morocco_matches["morocco_result"] == "Win").sum())
print("Draws:", (morocco_matches["morocco_result"] == "Draw").sum())
print("Losses:", (morocco_matches["morocco_result"] == "Loss").sum())

print("Goals scored by Morocco:", morocco_matches["morocco_goals"].sum())
print("Goals conceded:", morocco_matches["opponent_goals"].sum())
print("Average goals scored per match:", round(morocco_matches["morocco_goals"].mean(), 2))
print("Average goals conceded per match:", round(morocco_matches["opponent_goals"].mean(), 2))
```

```python
matches_per_year_morocco = morocco_matches.groupby("year").size()
wins_per_year_morocco = morocco_matches[morocco_matches["morocco_result"] == "Win"].groupby("year").size()

plt.figure(figsize=(12, 5))
matches_per_year_morocco.plot(label="Matches played")
wins_per_year_morocco.plot(label="Wins")
plt.title("Morocco - Matches and Wins Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
morocco_goals_by_year = morocco_matches.groupby("year")["morocco_goals"].mean()
opponent_goals_by_year = morocco_matches.groupby("year")["opponent_goals"].mean()

plt.figure(figsize=(12, 5))
morocco_goals_by_year.plot(label="Morocco avg goals")
opponent_goals_by_year.plot(label="Opponent avg goals")
plt.title("Morocco - Average Goals Scored vs Conceded by Year")
plt.xlabel("Year")
plt.ylabel("Average goals")
plt.legend()
plt.show()
```

```python
decade_summary = morocco_matches.groupby("decade").agg(
    matches=("morocco_result", "count"),
    wins=("morocco_result", lambda x: (x == "Win").sum()),
    draws=("morocco_result", lambda x: (x == "Draw").sum()),
    losses=("morocco_result", lambda x: (x == "Loss").sum()),
    avg_goals_scored=("morocco_goals", "mean"),
    avg_goals_conceded=("opponent_goals", "mean")
).reset_index()

decade_summary["win_rate"] = round((decade_summary["wins"] / decade_summary["matches"]) * 100, 2)

decade_summary
```

```python
plt.figure(figsize=(12, 5))
plt.bar(decade_summary["decade"].astype(str), decade_summary["win_rate"])
plt.title("Morocco - Win Rate by Decade")
plt.xlabel("Decade")
plt.ylabel("Win rate (%)")
plt.show()
```

```python
morocco_by_tournament = morocco_matches.groupby("tournament").agg(
    matches=("morocco_result", "count"),
    wins=("morocco_result", lambda x: (x == "Win").sum()),
    draws=("morocco_result", lambda x: (x == "Draw").sum()),
    losses=("morocco_result", lambda x: (x == "Loss").sum()),
    goals_scored=("morocco_goals", "sum"),
    goals_conceded=("opponent_goals", "sum")
).reset_index()

morocco_by_tournament["win_rate"] = round((morocco_by_tournament["wins"] / morocco_by_tournament["matches"]) * 100, 2)

morocco_by_tournament.sort_values("matches", ascending=False).head(15)
```

```python
top_tournaments = morocco_by_tournament.sort_values("matches", ascending=False).head(10)

plt.figure(figsize=(12, 5))
plt.bar(top_tournaments["tournament"], top_tournaments["matches"])
plt.title("Morocco - Top 10 Tournaments by Number of Matches")
plt.xlabel("Tournament")
plt.ylabel("Matches")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
morocco_matches["venue_type"] = morocco_matches["morocco_is_home"].map({True: "Home", False: "Away"})

venue_summary = morocco_matches.groupby("venue_type").agg(
    matches=("morocco_result", "count"),
    wins=("morocco_result", lambda x: (x == "Win").sum()),
    draws=("morocco_result", lambda x: (x == "Draw").sum()),
    losses=("morocco_result", lambda x: (x == "Loss").sum()),
    avg_goals_scored=("morocco_goals", "mean"),
    avg_goals_conceded=("opponent_goals", "mean")
).reset_index()

venue_summary["win_rate"] = round((venue_summary["wins"] / venue_summary["matches"]) * 100, 2)

venue_summary
```

```python
plt.figure(figsize=(7, 4))
plt.bar(venue_summary["venue_type"], venue_summary["win_rate"])
plt.title("Morocco - Home vs Away Win Rate")
plt.xlabel("Venue")
plt.ylabel("Win rate (%)")
plt.show()
```
